```python
# dist = dist[dist["stockChange"] > 0]
dist = dist[dist["vinstmarginal"] > -1000]

dist = dist[dist["bruttomarginal"] > -1000]
dist = dist[dist["bruttomarginal"] < 1000]

dist = dist[dist["operativtkassaflodesmarginal"] > -2000]
dist = dist[dist["operativtkassaflodesmarginal"] < 2000]

dist = dist[dist["omsattningstillvaxt"] < 8000]

dist = dist[dist["vinst/aktie"] < 100]

dist = dist[dist["egetkapital/aktie"] < 600]
dist = dist[dist["egetkapital/aktie"] > -200]

dist = dist[dist["fcf/aktie"] < 200]
dist = dist[dist["fcf/aktie"] > -200]

dist = dist[dist["rorelsemarginal"] < 500]

dist = dist[dist["ebita/aktie"] < 55]
dist = dist[dist["ebita/aktie"] > -55]

dist = dist[dist["operativkassaflode/aktie"] < 55]
dist = dist[dist["operativkassaflode/aktie"] > -55]

dist = dist[dist["aretskassaflode/aktie"] < 60]
dist = dist[dist["aretskassaflode/aktie"] > -60]

dist = dist[dist["vinstmarginal"] < 1000]


for column in dist:
    if "stockChange" not in column: 
        plt.figure(figsize=(20,20))
        sns.jointplot(column, "stockChangePct", data=dist)
        plt.show()
```

```python
dist["bull"] = dist["stockChangePct"] > 0

prognos = dist[["bruttomarginal", "operativtkassaflodesmarginal", "omsattningstillvaxt", "vinst/aktie", "egetkapital/aktie", "fcf/aktie", "rorelsemarginal","ebita/aktie","ebit/aktie","operativkassaflode/aktie","aretskassaflode/aktie","vinstmarginal","bull"]]
prognos.head()
```

```python
#corelation matrix.
corr= prognos.corr()
mask = np.array(corr)
mask[np.tril_indices_from(mask)] = False
fig=plt.gcf()
fig.set_size_inches(12,10)
sns.heatmap(data=corr,mask=mask,square=True,annot=True,cbar=True,cmap=plt.cm.Reds)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix


X = prognos[["bruttomarginal", "operativtkassaflodesmarginal", "omsattningstillvaxt", "vinst/aktie", "egetkapital/aktie", "fcf/aktie", "rorelsemarginal","ebita/aktie","ebit/aktie","operativkassaflode/aktie","aretskassaflode/aktie","vinstmarginal"]]
X = preprocessing.scale(X)
Y = prognos[["bull"]]

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier(loss="hinge", penalty="l2", max_iter=100)
clf.fit(x_train, y_train)
predictions = clf.predict(x_test)
print(accuracy_score(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

```python
from sklearn import svm
clf = svm.SVC(gamma='scale')
clf.fit(x_train, y_train)
predictions = clf.predict(x_test)
print(accuracy_score(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

```python
from tensorflow.python.client import device_lib

device_lib.list_local_devices()
```

```python
from keras.models import Sequential
from keras import layers
 
input_dim = x_train.shape[1]  # Number of features

model = Sequential()
model.add(layers.Dense(24, input_dim=input_dim, activation='relu'))
model.add(layers.Dense(48, activation='relu'))
# model.add(layers.Dropout(.8))

model.add(layers.Dense(96, activation='relu'))
# model.add(layers.Dropout(.8))

model.add(layers.Dense(72, activation='relu'))
# model.add(layers.Dropout(.8))
model.add(layers.Dense(48, activation='relu'))

model.add(layers.Dense(24, activation='relu'))

model.add(layers.Dense(12, activation='relu'))

model.add(layers.Dense(1, activation='sigmoid'))


```

```python
model.compile(loss='binary_crossentropy', 
               optimizer='adam', 
               metrics=['accuracy'])
model.summary()
```

```python
from keras.callbacks import EarlyStopping

es = EarlyStopping(monitor='val_loss', mode='min', patience=10)

model_output = model.fit(x_train, y_train,
                    epochs=1000,
                    verbose=10,
                    validation_data=(x_test, y_test),
                    batch_size=10,
                    callbacks=[es])
```

```python
loss, accuracy = model.evaluate(x_train, y_train, verbose=False)
print("Training Accuracy: {:.4f}".format(accuracy))
loss, accuracy = model.evaluate(x_test, y_test, verbose=False)
print("Testing Accuracy:  {:.4f}".format(accuracy))


print('Training Accuracy : ' , np.mean(model_output.history["acc"]))
print('Validation Accuracy : ' , np.mean(model_output.history["val_acc"]))



# Plot training & validation accuracy values
plt.plot(model_output.history['acc'])
plt.plot(model_output.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(model_output.history['loss'])
plt.plot(model_output.history['val_loss'])
plt.title('model_output loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
